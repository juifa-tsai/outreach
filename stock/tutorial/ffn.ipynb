# ffn取得股價
- 底層以yahoo api取得股價

```python
%pylab inline
import ffn
import pandas as pd

plt.rcParams['axes.unicode_minus'] = False
```

```python
 ffn.get?
```

```python
start='2018-01-01'
prices = ffn.get('2330.tw, 3008.tw, 2324.tw, 2337.tw, 1102.tw', start=start)
prices.head()
```

# 下跌幅度

```python
ffn.to_drawdown_series(prices).plot(figsize=(15,10))
```

# 分析歷史數據分析

```python
perf = prices.calc_stats()
perf.plot()
```

```python
perf.display()
```

```python
perf.stats
```

```python
returns = prices.to_returns().dropna()
print(returns.head())
ax = returns.hist(figsize(10, 5))
```
