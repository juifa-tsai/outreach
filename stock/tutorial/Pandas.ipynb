# 成績範例

```python
import pandas as pd
score = pd.read_csv('成績.csv', index_col = 0, engine='python') #未添加engine='python'，pandas無法讀取中文名稱檔案
score.head()
```

```python
import pandas as pd
with open('成績.csv') as f:
    score = pd.read_csv(f)
score.set_index('座號')  
score.head()
```

# 過濾、篩選

```python
score[score['國文']>60]
```

```python
score[(score['數學']>60) & (score['英文']>60)]
```

# 取最大值與最小值

```python
#國文成績最好
score['國文'].max()
```

```python
score.loc[[score['國文'].idxmax()]]
```

```python
#國文成績最差
score['國文'].min()
```

```python
score.loc[[score['國文'].idxmin()]]
```

```python
score['國文'].min()
```

# 總分計算

```python
score1 = score
score1['總分'] = score1['國文'] + score1['英文'] + score1['數學'] + score1['地理'] + score1['歷史'] + score1['理化'] + score1['電腦']
score1.head()
```

```python
score1 = score
score1['總分'] = score1.apply(lambda x: sum(x[1:8]), axis = 1)
score1.head()
```

# 成績平均

```python
from statistics import mean 
score1 = score
score1['平均'] = score1.apply(lambda x: '{:.2f}'.format(mean(x[1:8])), axis = 1)
score1.head()
```

# 成績排名

```python
score1['總排名'] = score1['總分'].rank(method='min', ascending=False)
score1 = score1.sort_values('總排名')
score1.head()
```

# 合併merge

```python
import numpy as np
import pandas as pd

left = pd.DataFrame({'key': ['K0', 'K1', 'K2', 'K3'],
                             'A': ['A0', 'A1', 'A2', 'A3'],
                             'B': ['B0', 'B1', 'B2', 'B3']})
right = pd.DataFrame({'key': ['K0', 'K1', 'K2', 'K3'],
                              'C': ['C0', 'C1', 'C2', 'C3'],
                              'D': ['D0', 'D1', 'D2', 'D3']})

print(left)
```

```python
print(right)
```

```python
res = pd.merge(left, right, on='key')
print(res)
```

```python
left = pd.DataFrame({'key1': ['K0', 'K0', 'K1', 'K2'],
                      'key2': ['K0', 'K1', 'K0', 'K1'],
                      'A': ['A0', 'A1', 'A2', 'A3'],
                      'B': ['B0', 'B1', 'B2', 'B3']})
right = pd.DataFrame({'key1': ['K0', 'K1', 'K1', 'K2'],
                       'key2': ['K0', 'K0', 'K0', 'K0'],
                       'C': ['C0', 'C1', 'C2', 'C3'],
                       'D': ['D0', 'D1', 'D2', 'D3']})

print(left)
```

```python
print(right)
```

# merge 合併方法中，how的使用
+ left:以左邊為主
+ right：以右邊為主
+ outer：不管key
+ inner：以相同key為主，配合on

```python
res = pd.merge(left, right, on=['key1', 'key2'], how='inner')
print(res)
```

```python
res = pd.merge(left, right, on=['key1', 'key2'], how='outer')
print(res)
```

```python
res = pd.merge(left, right, on=['key1', 'key2'], how='left')
print(res)
```

```python
res = pd.merge(left, right, on=['key1', 'key2'], how='right')
print(res)
```

```python
left = pd.DataFrame({'A': ['A0', 'A1', 'A2'],
                     'B': ['B0', 'B1', 'B2']},
                     index=['K0', 'K1', 'K2'])
right = pd.DataFrame({'C': ['C0', 'C2', 'C3'],
                      'D': ['D0', 'D2', 'D3']},
                     index=['K0', 'K2', 'K3'])

print(left)
```

```python
print(right)
```

```python
res = pd.merge(left, right, left_index=True, right_index=True, how='outer')
print(res)
```

```python
res = pd.merge(left, right, left_index=True, right_index=True, how='inner')
print(res)
```

# 產生日期序列
### [Time Series / Date functionality](https://pandas.pydata.org/pandas-docs/stable/timeseries.html)
### ![Offset Aliases](./pandas_date_range.png "Offset Aliases")

生成日期序列
+ pandas提供pd.data_range()和pd.period_range()兩個方法，給定參數有起始時間、結束時間、生成時期的數目及時間頻率（freq='M'月，'D'天，'W '，週，'Y'年）等。
+ 兩種主要區別在於pd.date_range()生成的是DatetimeIndex格式的日期序列；pd.period_range()生成的是PeriodIndex格式的日期序列。

```python
import pandas as pd
df = pd.date_range("2017-10-01", periods=30, freq = 'D')
df
```

```python
import pandas as pd
df = pd.period_range("2017-10-01", periods=30, freq = 'M')
df
```
