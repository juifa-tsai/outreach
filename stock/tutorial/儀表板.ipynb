```python
import requests
from bs4 import BeautifulSoup as bs
import pandas as pd

stockid = input("請輸入股票代碼：")
url = "http://jsjustweb.jihsun.com.tw/z/zc/zcr/zcr_{}.djhtm".format(stockid)
res = requests.get(url)
soup = bs(res.text, 'lxml')
tb = soup.select('table')[2]
df = pd.read_html(tb.prettify('utf-8'), encoding='utf-8', header = 1)
df[0]
```

```python
pd.set_option("display.max_rows", None)
df = df[0][(df[0]['期別'] != '期別') & (df[0]['期別'] != '種類') & (df[0]['期別'].str.find('：') == -1)].dropna()
df = df.set_index('期別')
df
```

```python
df0 = df.T
df0 = df0.sort_index()
df0 = df0.apply(pd.to_numeric, errors='ignore')
df0
```

```python
%matplotlib inline
#%pylab inline
import matplotlib.pyplot as plt
import platform

plt.rcParams['axes.unicode_minus'] = False

def plot_fig(df, ax, title, alpha, fontsize = 14):    
    ax.set_title(title, fontsize=fontsize)
    df.plot(ax=ax, alpha = alpha)
    ax.legend(fontsize=fontsize)
    
#依據作業系統設定字型
sysstr = platform.system()
if(sysstr =="Windows"):
    if (platform.release == '7'):
        plt.rcParams['font.sans-serif'] = 'Microsoft YaHei'
    elif (platform.release == '8' or platform.release == '8.1'):
        plt.rcParams['font.sans-serif'] = 'SimHei'
    else:
        plt.rcParams['font.sans-serif'] = 'SimHei'
elif(sysstr == "Linux"):
    print ("Call Linux tasks")
elif(sysstr == "Darwin"):
    print ("Call Mac os tasks")    
else:
    print ("Other System tasks")    
    
df0['應收帳款週天數'] = 90 /df0['應收帳款週轉次']
df0['存貨週轉天數'] = 90 /df0['存貨週轉率(次)']
df0['營運周轉天數'] = df0['應收帳款週天數'] + df0['存貨週轉天數']

fig, axes = plt.subplots(3, 3, figsize = (18, 12))
fig.suptitle(str(stockid) + '財報', fontsize=20)

plot_fig(df0[['營業毛利率', '營業利益率', '稅前淨利率', '稅後淨利率']], axes[0,0], '毛利率 & 淨利率', 0.5)
plot_fig(df0[['業外收支/營收']], axes[0,1], '業外收支/營收', 0.5)
plot_fig(df0[['ROE(A)─稅後']], axes[0,2], 'ROE', 0.5)

plot_fig(df0[['每股盈餘']], axes[1,0], '每股盈餘', 0.5)
plot_fig(df0[['每股淨值(F)(TSE公告數)']], axes[1,1], '每股淨值', 0.5)
plot_fig(df0[['應收帳款週天數', '存貨週轉天數', '營運周轉天數']], axes[1,2], '營運周轉天數', 0.5)

plot_fig(df0[['負債比率％']], axes[2,0], '負債比率％', 0.5)
plot_fig(df0[['流動比率', '速動比率']], axes[2,1], '流動比率 & 速動比率', 0.5)
plot_fig(df0[['利息保障倍數']], axes[2,2], '利息保障倍數', 0.5)
```

```python
%matplotlib inline
#%pylab inline
import matplotlib.pyplot as plt
import platform
import json

plt.rcParams['axes.unicode_minus'] = False 

#依據作業系統設定字型
sysstr = platform.system()
if(sysstr =="Windows"):
    if (platform.release == '7'):
        plt.rcParams['font.sans-serif'] = 'Microsoft YaHei'
    elif (platform.release == '8' or platform.release == '8.1'):
        plt.rcParams['font.sans-serif'] = 'SimHei'
    else:
        plt.rcParams['font.sans-serif'] = 'SimHei'
elif(sysstr == "Linux"):
    print ("Call Linux tasks")
elif(sysstr == "Darwin"):
    print ("Call Mac os tasks")    
else:
    print ("Other System tasks")   
    
row = col = 3

df0['應收帳款週天數'] = 90 /df0['應收帳款週轉次']
df0['存貨週轉天數'] = 90 /df0['存貨週轉率(次)']
df0['營運周轉天數'] = df0['應收帳款週天數'] + df0['存貨週轉天數']

chartconfigure = [
    {
        'title': '毛利率與淨利率',
        'x' : ['期別'],
        'y' : ['營業毛利率', '營業利益率', '稅前淨利率', '稅後淨利率'],
        'charttype' : ['line', 'line', 'line', 'line'],
        'fontsize' : 20,
        'alpha' : 0
       },
    {
        'title': '業外收支/營收',
        'x' : ['期別'],
        'y' : ['業外收支/營收'],
        'charttype' : ['line'],
        'fontsize' : 20,
        'alpha' : 0
       },
    {
        'title': 'ROE',
        'x' : ['期別'],
        'y' : ['ROE(A)─稅後'],
        'charttype' : ['bar'],
        'fontsize' : 20,
        'alpha' : 0
       },
    {
        'title': '每股盈餘',
        'x' : ['期別'],
        'y' : ['每股盈餘'],
        'charttype' : ['bar'],
        'fontsize' : 20,
        'alpha' : 0
       },
    {
        'title': '每股淨值',
        'x' : ['期別'],
        'y' : ['每股淨值(F)(TSE公告數)'],
        'charttype' : ['line'],
        'fontsize' : 20,
        'alpha' : 0
       },
    {
        'title' : '營運周轉天數',
        'x' : ['期別'],
        'y' : ['應收帳款週天數', '存貨週轉天數', '營運周轉天數'],
        'charttype' : ['line', 'line', 'bar'],
        'fontsize' : 20,
        'alpha' : 0
       },
    {
        'title': '負債比率％',
        'x' : ['期別'],
        'y' : ['負債比率％'],
        'charttype' : ['line'],
        'fontsize' : 20,
        'alpha' : 0
       },
    {
        'title': '流動比率 & 速動比率',
        'x' : ['期別'],
        'y' : ['流動比率', '速動比率'],
        'charttype' : ['line', 'line'],
        'fontsize' : 20,
        'alpha' : 0
       },
    {
        'title': '利息保障倍數',
        'x' : ['期別'],
        'y' : ['利息保障倍數'],
        'charttype' : ['line'],
        'fontsize' : 20,
        'alpha' : 0
       },
]

fig, axes = plt.subplots(row, col, figsize = (20, 16))
fig.suptitle(str(stockid) + '財報', fontsize=25)
fig.subplots_adjust(left=0.0,bottom=0.0, hspace=0.55, wspace=0.35)

for i in range(len(chartconfigure)):
    jd = json.loads(json.dumps(chartconfigure[i]))
    ax = axes[int(i/row), i%col]
    for seq in jd['y']:
        if jd['alpha']:
            df0[seq].plot(ax=ax, kind = jd['charttype'][jd['y'].index(seq)], alpha = jd['alpha']) 
        else:
            df0[seq].plot(ax=ax, kind = jd['charttype'][jd['y'].index(seq)]) 
    ax.set_title(jd['title'], fontsize=jd['fontsize'])
    ax.legend(fontsize=jd['fontsize'])
    ax.tick_params(axis='x', labelsize=jd['fontsize'])
    ax.tick_params(axis='y', labelsize=jd['fontsize'])
    ax.set_xticklabels(df0.index, rotation=90)
```

plt.subplots_adjust(bottom=0.08, top=0.92, left=0.10, right=0.95, hspace=0.25, wspace=0.35)

一幅圖稱為figure, 其繪畫區域稱為axes:
bottom, top: 即 axes距離畫板底部的距離 (畫板的高度取1)
left, right: 即 axes距離畫板左邊的距離 (畫板的寬度取1)
hspace: hight space 上下axes的距離
wspace: width space 左右axse的距離

注: bottom, top, left, right 不管figure實際長度和寬度為多少,都會歸一化為1,這裡填的數值,更確切的說是`占的比例`

```python
import requests
from bs4 import BeautifulSoup as bs
import pandas as pd

stockid = 2330
url = "http://jsjustweb.jihsun.com.tw/z/zc/zcr/zcr_{}.djhtm"
url = url.format(stockid)
res = requests.get(url)
soup = bs(res.text, 'lxml')
tb = soup.select('table')[2]
df0 = pd.read_html(tb.prettify('utf-8'), encoding='utf-8', header = 1)
pd.set_option("display.max_rows", None)
pd.set_option("display.max_columns", None)
df0 = df0[0][(df0[0]['期別'] != '期別') & (df0[0]['期別'] != '種類') & (df0[0]['期別'].str.find('：') == -1)].dropna()
df0 = df0.set_index('期別')
df = df0.T
df = df.sort_index()
#df = df.apply(pd.to_numeric, errors='ignore')
df
```

```python
import webbrowser
from pyecharts import Line, Bar, Overlap, Grid

df['應收帳款天數'] = list(map(lambda x: "{:.2f}".format(90 / x), df['應收帳款週轉次'].astype(float)))
df['存貨週轉天數'] = list(map(lambda x: "{:.2f}".format(90 / float(x)), df['存貨週轉率(次)']))
df['營運周轉天數'] = list(map(lambda x,y: "{:.2f}".format(x+y), df['應收帳款天數'].astype(float), df['存貨週轉天數'].astype(float)))
attr = df.index

#
#毛利率與淨利率 圖屬性右下，實際位置左上
#
line1 = Line("毛利率與淨利率")
line1.add("營業毛利率", 
          attr, 
          df['營業毛利率'], 
          #mark_point=["max", "min"],
          mark_line=["average"],
          is_smooth=True,       #平滑曲線設定
         )

line1.add("營業利益率", 
          attr, 
          df['營業利益率'], 
          #mark_point=["max", "min"],
          mark_line=["average"],
         )

line1.add("稅前淨利率", 
          attr, 
          df['稅前淨利率'], 
          #mark_point=["max", "min"],
          #mark_line=["average"],
         )

line1.add("稅後淨利率", 
          attr, 
          df['稅後淨利率'], 
          #mark_point=["max", "min"],
          #mark_line=["average"],
          legend_pos="30%",
          legend_orient='vertical'
         )

#
#每股盈餘 圖屬性左下，實際位置右上
#
line2 = Line(title='每股盈餘', title_pos="50%")
line2.add('每股盈餘', 
          attr, 
          df['每股盈餘'], 
          #mark_point=["max", "min"],
          #mark_line=["average"],
          legend_pos="80%"
         )

#
#流動比與速動比 圖屬性右上，實際位置左下
#
line3 = Line("流動比與速動比", title_top="50%")
line3.add("流動比率", 
          attr, 
          df['流動比率'], 
          mark_point=["max", "min"],
          mark_line=["average"],
         )

line3.add("速動比率", 
          attr, 
          df['速動比率'],
          mark_point=["max", "min"],
          mark_line=["average"],
          legend_top="50%",
          legend_pos="20%"
         )

#
#營運周轉天數 圖屬性左上，實際位置右下
#
bar = Bar(title="營運周轉天數", title_top="50%", title_pos="50%", background_color = '#606060')
bar.add("營運周轉天數", 
        attr, 
        df['營運周轉天數'], 
        mark_point=["max", "min"],
        legend_top="50%", 
        legend_pos="80%",)
line4 = Line()
line4.add("應收帳款天數", 
          attr, 
          df['應收帳款天數'], 
          #mark_point=["max", "min"],
          #mark_line=["average"],
         )
line5 = Line()
line5.add("存貨週轉天數", 
          attr, 
          df['存貨週轉天數'], 
          #mark_point=["max", "min"],
          #mark_line=["average"],
         )

overlap = Overlap()
overlap.add(bar)
overlap.add(line4)
overlap.add(line5)

grid = Grid(height=800, width=1000)

grid.add(line1, grid_bottom="60%", grid_right="60%")

grid.add(overlap, grid_top ="60%", grid_left="60%")

grid.add(line2, grid_bottom="60%", grid_left="60%")

grid.add(line3, grid_top="60%", grid_right="60%")

grid.render('grid.html')
#webbrowser.open('grid.html')
grid
```
